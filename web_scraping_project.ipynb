```python
#17/10/2021
#import libraries
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
import os
```

```python
#Chossing the job title and its details
key_word = str(input("Please Enter The Job You Want >>> "))
print(f"Searching For {key_word} Jops")
page_num = 0
jop_title = []
company_name = []
location = []
skill = []
requirments = []
detalis = []
salary = []
links = []
directory = os.getcwd()
directory = directory + f"\{key_word}_Jops.csv"
flag = 0
flag_2 = 0
```

```python
while True:
    try:
        key_word = key_word.replace(" ", "+")
        main_link = f"https://wuzzuf.net/search/jobs/?a=hpb&q={key_word}&start={page_num}"
        print(f"Page Number {page_num} Searching...\nLink: {main_link}")
        result = requests.get(main_link)
        scr = result.content
        soup = BeautifulSoup(scr, "lxml")
        jop_limit = soup.find("strong").text.replace(" ", "").replace(",", "")
        page_limit = round(int(jop_limit) / 15)
        jop_titles = soup.find_all("a", {"target": "_blank", "rel": "noreferrer", "class": "css-o171kl"})
        company_names = soup.find_all("a", {"target": "_blank", "rel": "noreferrer", "class": "css-17s97q8"})
        locations = soup.find_all("span", {"class": "css-5wys0k"})
        skills = soup.find_all("div", {"class": "css-y4udm8"})
        for i in range(len(jop_titles)):
            jop_title.append(jop_titles[i].text)
            links.append("https://wuzzuf.net/" + jop_titles[i].get("href"))
            company_name.append(company_names[i].text[:-1])
            location.append(locations[i].text)
            skill.append(skills[i].text.replace("Â·", "|"))
        if page_num >= page_limit:
            break
        else:
            page_num += 1
    except:
        print("Error Occurs")
```

```python
for link in links:
    try:
        result = requests.get(link)
        scr = result.content
        soup = BeautifulSoup(scr, "lxml")
        detalis = soup.find("div", {"class": "matching-requirement-icon-container", })
        salary.append(detalis.text.strip())
        try:
            requirments.append(soup.find("span", {"itemprop": "responsibilities"}).ul.text.strip())

        except:
            requirments.append(soup.find("div", {"class": "css-1t5f0fr"}))
        flag += 1
        print(f"{flag} Jop Found")
    except:
        flag_2 += 1
        requirments.append("GO to Link")
        salary.append("GO to Link")
        print(f"{flag_2} Denied Pages from {len(links)} Page\nCheck Link: {link}\n{flag} Accepted Page")

data = [jop_title, company_name, location, salary, skill, requirments, links]
data_list = zip_longest(*data)
try:
    with open("D:\CIC\8th sem\Embedded Systems", "w", newline='') as jops:
        wr = csv.writer(jops)
        wr.writerow(["Jop Title", "Company Name", "Location", "Salary", "Skills", "Requires", "More Info"])
        wr.writerows(data_list)
        print(f"Done WebScraping\nFound {len(jop_title)} Jops\nGOOD LUCK -_*")
except:
    with open("D:\CIC\8th sem\Embedded Systems", "w", encoding="utf-8", newline='') as jops:
        wr = csv.writer(jops)
        wr.writerow(["Jop Title", "Company Name", "Location", "Salary", "Skills", "Requires", "More Info"])
        wr.writerows(data_list)
        print(f"Done WebScraping\nFound {len(jop_title)} Jops\nGOOD LUCK -_*")
        print("\nNOTE: UTF-8 CODED CAUSE OF NOT ENGLISH LITTERS")
```
